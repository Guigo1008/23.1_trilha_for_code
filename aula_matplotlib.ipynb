```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x: float) -> float:
  return x**2 + 3
def f2(x: float) -> float:
  return x**3
```

```python
x = np.linspace(-100, 100, 1000)
y2 = f2(x)
y = f(x)
```

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.title("Plotando a função x**2 + 3")
plt.xlabel("Plotando o x")
plt.ylabel("Plotando o y")
plt.grid()
plt.show()
```
